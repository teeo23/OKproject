```python
df['smokes'] = df['smokes'].map({'yes' : 1,
                                 'sometimes': 2,
                                 'when drinking': 3,
                                 'trying to quit' : 4,
                                 'no' : 5})
```

```python
df['drinks'] = df['drinks'].map({'desperately' : 1,
                                'very often' : 2,
                                'often' : 3,
                                'socially' : 4,
                                'rarely' : 5,
                                'not at all' : 6})
```

```python
new_df = df[['signs_clean', 'smokes', 'drinks']]
print(new_df)
new_df.shape
```

```python
new_df = new_df[new_df['signs_clean'].notna()]
```

```python
new_df.isna().any()
```

```python
new_df.fillna({'smokes':5,
           'drinks':6},
          inplace=True)
new_df.isna().any()
new_df.shape
print(new_df)
```

```python
from sklearn.preprocessing import scale

labels = new_df['signs_clean']
data = new_df[['smokes','drinks']]

scaled_data = scale(data, axis = 0)

print(scaled_data[0])
```

```python


from sklearn.model_selection import train_test_split
train_data, test_data, train_labels, test_labels = train_test_split(scaled_data, labels, test_size = 0.2, random_state = 1)

```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors = 100)
knn_model.fit(train_data, train_labels)
knn_predictions = knn_model.predict(train_data)
```

```python
print(knn_model.score(train_data, train_labels))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(knn_model, train_data, train_labels, xticks_rotation='vertical')
plt.savefig('Confusion matrix KNN')
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_class = DecisionTreeClassifier()
tree_class.fit(train_data, train_labels)
```

```python
print(tree_class.score(train_data, train_labels))
prediction = tree_class.predict(test_data)
```

```python
from sklearn.metrics import confusion_matrix

label = tree_class.classes_
print(label)
matrix = confusion_matrix(test_labels, prediction)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(tree_class, train_data, train_labels, xticks_rotation='vertical')
plt.savefig('Confusion Matrix Decision Tree')
plt.show()
```

```python
from sklearn.svm import SVC
svc_classifier = SVC(kernel = 'rbf', gamma = 0.5, C =2)
svc_classifier.fit(train_data, train_labels)
```

```python
print(svc_classifier.score(train_data, train_labels))
prediction = svc_classifier.predict(test_data)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(svc_classifier, train_data, train_labels, xticks_rotation='vertical')
plt.savefig('Support Vector Classifier ')
plt.show()
```

```python
cols = ['body_type', 'diet', 'orientation', 'education', 'religion_clean',
       'sex', 'job', 'signs_clean']
profile_df = df[cols].dropna()
profile_df.shape
```

```python
for column in cols[:-1]:
    profile_df = pd.get_dummies(profile_df, columns=[column], prefix = [column])
```

```python
profile_df.head()
```

```python
col_length = len(profile_df.columns)
X = profile_df.iloc[:, 1:col_length]
y = profile_df.iloc[:, 0:1]

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 0)

y_train = y_train.to_numpy().ravel()
y_test = y_test.to_numpy().ravel()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors = 5)
knn_model.fit(X_train, y_train)
knn_predictions = knn_model.predict(X_train)
```

```python
print(knn_model.score(X_train,y_train))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(knn_model, X_test, y_test, xticks_rotation='vertical')
plt.savefig('Confusion matrix KNN 2')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
print(classifier.score(X_train, y_train))
prediction = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

label = classifier.classes_
print(label)
matrix = confusion_matrix(y_test, prediction)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(classifier, X_test, y_test, xticks_rotation='vertical')
plt.savefig('Confusion Matrix Decision Tree 2')
plt.show()
```

```python
from sklearn.svm import SVC
svc_class = SVC(kernel = 'rbf', gamma = 0.5, C =2)
svc_class.fit(X_train,y_train)
```

```python
print(svc_class.score(X_train, y_train))
prediction = svc_class.predict(X_test)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(svc_class, X_test, y_test, xticks_rotation='vertical')
plt.savefig('Support Vector Classifier 2')
plt.show()
```
