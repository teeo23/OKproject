```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('profiles.csv')
print(df.head(10))
```

```python
df.columns
```

```python
print('The average age in the dataset is : ' , np.average(df['age']))
```

```python
sns.set_palette("Set2")
sns.countplot(data=df, y='sex')

plt.savefig('Ratio of Male & Female Members')
```

```python
sns.displot(data=df, x="age", hue="sex", kind="hist", binwidth = 5, multiple = "stack");
plt.savefig('Age comparison')
```

```python
sns.countplot(data=df, y='orientation', hue='sex')
plt.savefig('Sexual orientation')
```

```python
plt.figure(figsize=(8,10))
sns.countplot(data=df, y='education', hue='sex')
plt.savefig('Education')
```

```python
plt.figure(figsize=(8,10))
sns.countplot(data=df, y='job', hue='sex')
plt.savefig('Job')
```

```python
print('Number of unique signs is: ' + str(df.sign.nunique()))

print('Unique sign counts: ' + str(df.sign.unique()))
```

```python
df['signs_clean'] = df.sign.str.split().str.get(0)

print(df.signs_clean.nunique())
print(df.signs_clean.unique())
```

```python
sns.countplot(data=df, y='signs_clean')
plt.savefig('Signs count cleaned')
```

```python
df.signs_clean.value_counts()
```

```python
sns.countplot(data=df, y='signs_clean', hue = 'sex')
```
